```python
%sql
-- Column mask (hide sensitive column from non-admins)
CREATE FUNCTION training_catalog.sales.mask_email(email STRING)
  RETURN CASE WHEN IS_ACCOUNT_GROUP_MEMBER('admins') THEN email ELSE '****' END;

ALTER TABLE training_catalog.sales.customers
  ALTER COLUMN email SET MASK training_catalog.sales.mask_email;
```
