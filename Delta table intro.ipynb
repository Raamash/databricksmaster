1. What table gets created? what is delta?

```python
# Create a richer DataFrame
df = spark.createDataFrame([
    (1, "Alice", "USA"),
    (2, "Bob", "UK"),
    (3, "Charlie", "India"),
    (4, "Diana", "Germany"),
    (5, "Ethan", "Brazil")
], ["id", "name", "country"])

# Write to Delta
df.write.format("delta").saveAsTable("my_delta_table")

df.show()
```

2. Default file format - delta.No need to write using delta in databricks. Other file formats are manadotoy to be specified.

```python
%sql
-- sample managed table
CREATE TABLE IF NOT EXISTS sample_managed_table (
  id INT,
  name STRING,
  country STRING
)
;
```

3. As soon as you specify a LOCATION, Databricks automatically treats it as an external table. You don’t need to add the keyword EXTERNAL.

```python
%sql
-- sample external table.
CREATE TABLE sample_external_table (
  id INT,
  name STRING,
  country STRING
)
LOCATION 's3 bucket';
```
