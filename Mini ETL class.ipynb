```python
# Create a richer DataFrame
data = spark.createDataFrame([
    (1, "Alice", "USA"),
    (2, "Bob", "UK"),
    (3, "Charlie", "India"),
    (4, "Diana", "Germany"),
    (5, "Ethan", "Brazil")
], ["id", "name", "country"])

# Write to Delta
data.write.parquet("")

# Read back
df = spark.read.format("delta").load("/tmp/delta-table")
df.show()
```

```python
%sql
-- Managed Table
CREATE TABLE managed_table (id INT, name STRING, country STRING)
USING DELTA;

-- External Table
CREATE TABLE external_table (id INT, name STRING, country STRING)
USING DELTA
LOCATION '/mnt/delta/external_table';
```

```python
# New data with extra column 'age'
new_data = spark.createDataFrame([
    (6, "Fiona", "Canada", 29),
    (7, "George", "Japan", 34)
], ["id", "name", "country", "age"])

# Append with schema evolution
new_data.write.format("delta").option("mergeSchema", "true").mode("append").save("/tmp/delta-table")
```

```python
%sql
-- Partitioned table by country
CREATE TABLE partitioned_table
USING DELTA
PARTITIONED BY (country)
AS SELECT * FROM managed_table;

-- Z-Ordering for query optimization
OPTIMIZE partitioned_table
ZORDER BY (id);
```
